```python
import pandas as pd
pd.options.display.max_columns = None
pd.set_option('display.max_rows', None)
```

```python
sic_table = pd.read_parquet("SicCodesAndStructure.parquet", engine='pyarrow')
```

```python
sic_table.columns = ["SIC", "SIC_Desc", "Level", "Parent", "Parent_Desc"]
```

```python
sic_table["Level"].describe()
```

```python
sic_table = sic_table[sic_table["Level"] > 2]
sic_table
```

```python
sic_table_by_sic = sic_table.set_index('SIC')
def find_root_parent(row, previous_parent):
    current_parent = row['Parent']
    if (row['Level'] <= 2):
        return current_parent
    
    return find_root_parent(sic_table_by_sic.loc[current_parent], current_parent)

for index, row in sic_table_by_sic.iterrows():
    sic_table_by_sic.at[index, 'root_parent'] = find_root_parent(row, row['Parent'])
```

```python
sic_table_by_sic.reset_index()
```

```python
sic_table_by_sic.to_excel("feature-eng-data/sic_level_1_mapping.xlsx")
```

```python
sic_table_by_sic = sic_table.set_index('SIC')
def find_root_parent(row, previous_parent):
    current_parent = row['Parent']
    if (row['Level'] <= 3):
        return current_parent
    
    return find_root_parent(sic_table_by_sic.loc[current_parent], current_parent)

for index, row in sic_table_by_sic.iterrows():
    sic_table_by_sic.at[index, 'parent_lv2'] = find_root_parent(row, row['Parent'])
```

```python
sic_table_by_sic  = sic_table_by_sic.drop(columns=["SIC_Desc", "Level", "Parent", "Parent_Desc"])
sic_table_by_sic = sic_table_by_sic.reset_index()
```
