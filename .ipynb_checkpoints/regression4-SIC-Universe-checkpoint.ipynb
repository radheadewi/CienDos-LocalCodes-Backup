Key findings within this notebook experiment:
1. The EstAssetsM, EstFTE, and EstTurnoverM are not good estimates because the values of companies in ScoresABC1C2DE are significantly very different.

```python
import pandas as pd
pd.options.display.max_columns = None
pd.set_option('display.max_rows', 20)
```

```python
companies_table = pd.read_parquet("feature-eng-data/companies_tablev2.parquet", engine='pyarrow')
companies_table.isna().sum()
```

```python
scores_table = pd.read_excel("feature-eng-data/ScoresABC1C2DE.xlsx")
scores_table = scores_table.drop(columns=["NatSic", "NatSicDesc", "AuditID", "ScoreD", "ScoreE"])
```

```python
scores_table_in_company = companies_table[companies_table["CompanyNumber"].isin(scores_table["Company_Number"])]
scores_table_in_company.isna().sum()
```

```python
scores_table_in_company["FTE"].head()
```

Now we are trying to fill in Assets, Turnover and FTE values with their estimation values from BaCoDaAll.

```python
companies_table['AssetsM'] = companies_table['AssetsM'].fillna(companies_table['EstAssetsM'])
companies_table['TurnoverM'] = companies_table['TurnoverM'].fillna(companies_table['EstTurnoverM'])
companies_table['FTE'] = companies_table['FTE'].fillna(companies_table['EstFTE'])
```

```python
companies_table.isna().sum()
```

```python
scores_table_in_company = companies_table[companies_table["CompanyNumber"].isin(scores_table["Company_Number"])]
scores_table_in_company.isna().sum()
```
